```python
avg=[]
grade=[]



def Topper():
    for i in range(0,1000):
        add=st['math_score'][i]+st['math_score'][i]+st['math_score'][i]
        avg.append(add)
#         float(avg)
#     [float(i) for i in avg]
        avg[i]=avg[i]/3
    
    print('The maximum marks scored in the class is',max(avg))
    
#     x=len(st['math_score'])



def result(avg):
    for marks in avg:
        if marks>90.0:
            grade.append('A+')
        elif marks>80.0 and marks<90.0:
            grade.append('B+')
        elif marks>70.0 and marks<80.0:
            grade.append('C+')
        elif marks>60.0 and marks<70.0:
            grade.append('D+')
        elif marks>50.0 and marks<60.0:
            grade.append('E+')
        else:
            grade.append('F')
    return grade
```
